```python
import pandas as pd

# First Task
df = pd.read_excel('sample_data.xlsx')
df
```

```python
# Task 2
df_avg = df['Narx'].mean()
df['Narx'] = df['Narx'].fillna(df_avg)
df
```

```python
# Task 3
def filter_function(value):
    if value < 0:
        return 0
    return value

df['Miqdor'] = df['Miqdor'].apply(filter_function)
df
```

```python
# Task 4
categories = df['Kategoriya'].unique()
categories
```

```python
# Task 5
df.to_csv('cleaned_data.xlsx')
```
